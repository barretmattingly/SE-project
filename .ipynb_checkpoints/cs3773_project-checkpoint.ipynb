```python
class Produce:
    def __init__ (self, id_, name_, price_, qty_):
        self.id = id_
        self.name = name_
        self.price = price_
        self.qty = qty_

def createProduce(id, name, price, qty):
    produceDic[id] = Produce(id, name, price, qty)

def updateProduce(produce):
    produceDic[produce.id] = produce

def deleteProduce(id):
    del produceDic[id]

def loadProduce():
    filepath = 'data/Produce.csv'
    file = open(filepath, 'r')
    for line in file.readlines():
        tmp = line.strip().split(',')
        createProduce(tmp[0],tmp[1],tmp[2],tmp[3])
    file.close()

def saveProduce():
    filepath = 'data/Produce.csv'
    file = open(filepath, 'w')
    for key in produceDic.keys():
        p = produceDic[key]
        tmp = p.id+','+p.name+','+p.price+','+p.qty
        file.write(tmp)
    file.close()
```
